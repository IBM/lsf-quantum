```python
import matplotlib.pyplot as plt
import numpy as np

# Data
samples = [100, 1000, 10000, 100000]
sequential = [81.99, 693, 1640, np.inf]

mpi_4_way = [34.98, 91.33, 91.08, 123.81]
mpi_8_way = [30.42, 224.73, 423.30, 438.46]

# Plot
plt.figure(figsize=(8, 5))

plt.plot(samples, sequential, marker='o', label='Sequential')
plt.plot(samples, mpi_4_way, marker='s', label='4-way MPI')
plt.plot(samples, mpi_8_way, marker='^', label='8-way MPI')

# Labels and title
plt.xlabel('Samples')
plt.ylabel('Time')
plt.title('Circuits Execution Time')

# Use log scale for X axis because samples vary by powers of 10
plt.xscale('log')

# Grid and legend
plt.grid(True, which='both', linestyle='--', linewidth=0.5)
plt.legend()

# Layout and display
plt.tight_layout()
#plt.show()
plt.savefig("mpi_performance_plot.png", dpi=300)
```
